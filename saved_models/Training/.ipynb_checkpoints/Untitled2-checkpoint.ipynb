```python
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
```

```python
# Set constants
IMAGE_SIZE = 224
BATCH_SIZE = 32
NUM_CLASSES = 4
```

```python
def _clahe_fn(image):
    image = image.numpy()
    lab = cv2.cvtColor(image, cv2.COLOR_RGB2LAB)
    l, a, b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    cl = clahe.apply(l)
    merged = cv2.merge((cl, a, b))
    final = cv2.cvtColor(merged, cv2.COLOR_LAB2RGB)
    return final

def apply_clahe(image):
    image = tf.cast(image * 255.0, tf.uint8)
    image = tf.numpy_function(func=_clahe_fn, inp=[image], Tout=tf.uint8)
    image.set_shape([None, None, 3])
    return image / 255.0
```

```python
resize_and_rescale = tf.keras.Sequential([
    layers.Resizing(IMAGE_SIZE, IMAGE_SIZE),
    layers.Rescaling(1./255)
])

data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.1)
])
```

```python
def load_dataset(Plantvillage):
    return tf.keras.preprocessing.image_dataset_from_directory(
        "PlantVillage",
        shuffle=True,
        image_size=(IMAGE_SIZE, IMAGE_SIZE),
        batch_size=BATCH_SIZE
    )

def get_dataset_partitions_tf(ds, train_split=0.8, val_split=0.1, test_split=0.1, shuffle=True, shuffle_size=10000):
    assert (train_split + test_split + val_split) == 1
    ds_size = len(ds)
    if shuffle:
        ds = ds.shuffle(shuffle_size, seed=12)
    train_size = int(train_split * ds_size)
    val_size = int(val_split * ds_size)
    train_ds = ds.take(train_size)
    val_ds = ds.skip(train_size).take(val_size)
    test_ds = ds.skip(train_size).skip(val_size)
    return train_ds, val_ds, test_ds
```

```python
def prepare_dataset(train_ds, val_ds, test_ds, use_clahe=True):
    if use_clahe:
        train_ds = train_ds.map(
            lambda x, y: (resize_and_rescale(apply_clahe(data_augmentation(x, training=True))), y),
            num_parallel_calls=tf.data.AUTOTUNE
        )
    else:
        train_ds = train_ds.map(
            lambda x, y: (resize_and_rescale(data_augmentation(x, training=True)), y),
            num_parallel_calls=tf.data.AUTOTUNE
        )
    val_ds = val_ds.map(lambda x, y: (resize_and_rescale(x), y), num_parallel_calls=tf.data.AUTOTUNE)
    test_ds = test_ds.map(lambda x, y: (resize_and_rescale(x), y), num_parallel_calls=tf.data.AUTOTUNE)
    return train_ds.prefetch(tf.data.AUTOTUNE), val_ds.prefetch(tf.data.AUTOTUNE), test_ds.prefetch(tf.data.AUTOTUNE)
```

```python
# Replace with the path to your dataset folder (must contain 4 subfolders for the classes)
dataset_dir = "PlantVillage"  # or your actual dataset path

full_dataset = load_dataset(dataset_dir)
train_ds, val_ds, test_ds = get_dataset_partitions_tf(full_dataset)
train_ds, val_ds, test_ds = prepare_dataset(train_ds, val_ds, test_ds, use_clahe=True)
```

```python
# ShuffleNetV2 unit
def shufflenet_unit(x, out_channels, strides):
    def channel_shuffle(x, groups=2):
        h, w, c = x.shape[1], x.shape[2], x.shape[3]
        x = tf.reshape(x, [-1, h, w, groups, c // groups])
        x = tf.transpose(x, [0, 1, 2, 4, 3])
        return tf.reshape(x, [-1, h, w, c])
    
    inp_channels = x.shape[-1]
    branch_channels = out_channels // 2
    
    if strides == 1:
        x1, x2 = tf.split(x, num_or_size_splits=2, axis=-1)
        out = tf.concat([x1, _conv_branch(x2, branch_channels, strides)], axis=-1)
    else:
        out = tf.concat([_conv_branch(x, branch_channels, strides),
                         _conv_branch(x, branch_channels, strides)], axis=-1)
    return channel_shuffle(out)

# Helper conv branch
def _conv_branch(x, out_channels, strides):
    x = layers.Conv2D(out_channels, 1, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.DepthwiseConv2D(3, strides=strides, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Conv2D(out_channels, 1, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    return x
```

```python
def build_shufflenetv2(input_shape=(224,224,3), num_classes=4, scale=1.0):
    out_channels = {
        0.5: [24, 48, 96, 192, 1024],
        1.0: [24, 116, 232, 464, 1024],
        1.5: [24, 176, 352, 704, 1024],
        2.0: [24, 244, 488, 976, 2048]
    }[scale]
    
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv2D(out_channels[0], 3, strides=2, padding='same', use_bias=False)(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.MaxPooling2D(3, strides=2, padding='same')(x)

    for reps, channels in zip([4, 8, 4], out_channels[1:4]):
        for i in range(reps):
            x = shufflenet_unit(x, channels, strides=2 if i == 0 else 1)

    x = layers.Conv2D(out_channels[-1], 1, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(num_classes, activation='softmax')(x)

    return models.Model(inputs, x)
```

```python
def shufflenet_unit(x, C_out, s):
    def shuffle(c):
        h, w = c.shape[1], c.shape[2]; ch = c.shape[3]
        c_ = tf.reshape(c, [-1, h, w, 2, ch//2])
        c_ = tf.transpose(c_, [0,1,2,4,3])
        return tf.reshape(c_, [-1, h, w, ch])
    
    def conv_branch(inp):
        y = layers.Conv2D(C_out//2,1,padding='same',use_bias=False)(inp)
        y = layers.BatchNormalization()(y)
        y = layers.ReLU()(y)
        y = layers.DepthwiseConv2D(3,strides=s,padding='same',use_bias=False)(y)
        y = layers.BatchNormalization()(y)
        y = layers.Conv2D(C_out//2,1,padding='same',use_bias=False)(y)
        y = layers.BatchNormalization()(y)
        y = layers.ReLU()(y)
        return y
    
    if s==1:
        x1, x2 = tf.split(x, 2, axis=-1)
        out = layers.Concatenate()([x1, conv_branch(x2)])
    else:
        out = layers.Concatenate()([conv_branch(x), conv_branch(x)])
    return shuffle(out)

def build_shufflenet(scale=1.0):
    cfg = {1.0:[24,116,232,464,1024]}[scale]
    inp = layers.Input((IMAGE_SIZE,IMAGE_SIZE,3))
    x = layers.Conv2D(cfg[0],3,2,'same',use_bias=False)(inp)
    x = layers.BatchNormalization()(x); x=layers.ReLU()(x)
    x = layers.MaxPooling2D(3,2,'same')(x)
    for reps, ch in zip([4,8,4], cfg[1:4]):
        for i in range(reps):
            x = shufflenet_unit(x, ch, s=2 if i==0 else 1)
    x = layers.Conv2D(cfg[-1],1,padding='same',use_bias=False)(x)
    x = layers.BatchNormalization()(x); x=layers.ReLU()(x)
    x = layers.GlobalAveragePooling2D()(x)
    return models.Model(inp, x)
```

```python
def woa_optimize(obj_fn, lb, ub, dim, pop=5, its=10):
    a0 = 2
    popX = [np.random.uniform(lb, ub) for _ in range(pop)]
    fit = [obj_fn(x) for x in popX]
    best = popX[np.argmin(fit)]
    best_fit = min(fit)

    for t in range(its):
        a = a0 - t * (a0 / its)
        for i in range(pop):
            r1, r2 = np.random.rand(), np.random.rand()
            A = 2 * a * r1 - a
            C = 2 * r2

            if random.random() < 0.5:
                D = abs(C * np.array(best) - np.array(popX[i]))
                popX[i] = best - A * D
            else:
                D = abs(best - popX[random.randint(0, pop - 1)])
                b, l = 1, random.uniform(-1, 1)
                popX[i] = D * np.exp(b * l) * np.cos(2 * np.pi * l) + best

            popX[i] = np.clip(popX[i], lb, ub)
        fit = [obj_fn(x) for x in popX]
        best = popX[np.argmin(fit)]
        best_fit = min(fit)

    return best
```

```python
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras import layers, models

# Parameters
IMAGE_SIZE = 224
NUM_CLASSES = 4

# Load feature extractor from TensorFlow Hub (ShuffleNet V2 1.0x)
feature_extractor_url = "https://tfhub.dev/google/imagenet/shufflenet_v2_1.0_224/feature_vector/5"
feature_extractor_layer = hub.KerasLayer(feature_extractor_url, input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3), trainable=False)

# Build the model
model = tf.keras.Sequential([
    layers.InputLayer(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3)),  # 💡 Ensures model is built
    feature_extractor_layer,
    layers.Dense(128, activation='relu'),
    layers.Dense(NUM_CLASSES, activation='softmax')
])

# Compile
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Print summary
model.summary()
```
